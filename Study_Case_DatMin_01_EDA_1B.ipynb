## 4. Dari data yang ada, apakah bisa dibuat rekomendasi sebaiknya RPPI membuat kantor cabang berikutnya di kota mana? Mengapa? (Asumsikan RPPI belum memiliki cabang di kota-kota yang disebutkan di data).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# DataFrame sudah ada, kita gunakan 'rumah' dari output sebelumnya

# Menghitung rata-rata harga per kota
rata_rata_harga = rumah.groupby('kota')['harga'].mean().reset_index()
rata_rata_harga.columns = ['kota', 'rata_rata_harga']

# Menghitung jumlah properti per kota
jumlah_properti = rumah['kota'].value_counts().reset_index()
jumlah_properti.columns = ['kota', 'jumlah_properti']

# Menggabungkan kedua DataFrame
rekomendasi = pd.merge(rata_rata_harga, jumlah_properti, on='kota')

# Menentukan rasio untuk potensi
rekomendasi['potensi'] = rekomendasi['jumlah_properti'] / rekomendasi['rata_rata_harga']

# Menampilkan rekomendasi
print(rekomendasi)

# Visualisasi rata-rata harga dan jumlah properti per kota
plt.figure(figsize=(12, 6))

# Bar plot untuk rata-rata harga
plt.subplot(1, 2, 1)
sns.barplot(x='kota', y='rata_rata_harga', data=rekomendasi, palette='viridis')
plt.title("Rata-rata Harga per Kota")
plt.xlabel("Kota")
plt.ylabel("Rata-rata Harga (Rp)")

# Bar plot untuk jumlah properti
plt.subplot(1, 2, 2)
sns.barplot(x='kota', y='jumlah_properti', data=rekomendasi, palette='viridis')
plt.title("Jumlah Properti per Kota")
plt.xlabel("Kota")
plt.ylabel("Jumlah Properti")

plt.tight_layout()
plt.show()

# Rekomendasi untuk lokasi kantor cabang
best_location = rekomendasi.loc[rekomendasi['potensi'].idxmax()]

print(f"\nRekomendasi lokasi kantor cabang RPPI: {best_location['kota']}")
print(f"Rata-rata harga: Rp {best_location['rata_rata_harga']:.2f}")
print(f"Jumlah properti: {best_location['jumlah_properti']}")
```

## 6. Jika RPPI ingin melakukan investasi (max) Rp. 25 Milyar minggu besok, apakah dari data anda dapat memberikan rekomendasi rumah mana saja yang akan berpotensi menghasilkan keuntungan bagi perusahaan?

### Apakah Luas Tanah dan Luas Bangunan mempengaruhi harga?

```python
# Scatter plot Harga vs LT (Luas Tanah) dan LB (Luas Bangunan)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='LT', y='harga', data=rumah)
plt.title('Harga vs Luas Tanah')
plt.xlabel('Luas Tanah (m2)')
plt.ylabel('Harga (Rp)')

plt.subplot(1, 2, 2)
sns.scatterplot(x='LB', y='harga', data=rumah)
plt.title('Harga vs Luas Bangunan')
plt.xlabel('Luas Bangunan (m2)')
plt.ylabel('Harga (Rp)')

plt.tight_layout()
plt.show()
```

Berdasarkan analisis dari dua scatter plot, baik Luas Tanah maupun Luas Bangunan memiliki pengaruh terhadap harga. Harga properti cenderung meningkat seiring dengan bertambahnya luas tanah dan luas bangunan. Pada grafik pertama (Harga vs Luas Tanah), terdapat hubungan positif antara luas tanah dan harga, meskipun ada variasi harga yang cukup besar untuk luas tanah yang sama, yang mengindikasikan bahwa faktor lain mungkin juga memengaruhi harga.

Demikian pula, pada grafik kedua (Harga vs Luas Bangunan), terlihat bahwa luas bangunan juga berpengaruh terhadap harga, dengan harga yang cenderung meningkat pada bangunan yang lebih besar. Namun, variabilitas harga juga tetap ada, khususnya pada bangunan dengan luas kecil hingga menengah.

Dengan demikian, dapat disimpulkan bahwa baik Luas Tanah maupun Luas Bangunan berperan dalam menentukan harga rumah, meski terdapat faktor lain yang mungkin turut berkontribusi pada fluktuasi harga.

### Apakah jenis sertifikat mempengaruhi harga

```python
plt.figure(figsize=(20, 8))
sns.barplot(x='sertifikat', y='harga', data=rumah, estimator=lambda x: sum(x) / len(x))
plt.title('Rata-rata Harga Berdasarkan Jenis Sertifikat')
plt.xlabel('Jenis Sertifikat')
plt.ylabel('Rata-rata Harga (Rp)')
plt.show()
```

Berdasarkan grafik batang yang menampilkan rata-rata harga berdasarkan jenis sertifikat, terlihat bahwa **jenis sertifikat properti mempengaruhi harga properti**.

- Sertifikat **SHM + IMB + PBB** memiliki harga rata-rata tertinggi, diikuti oleh **SHM + IMB** dan **SHM + Hak Pakai**, yang menunjukkan bahwa properti dengan kombinasi sertifikat ini umumnya dihargai lebih tinggi.
- Properti dengan sertifikat **SHM (Sertifikat Hak Milik)** dan **SHGB (Sertifikat Hak Guna Bangunan)** cenderung memiliki harga lebih rendah dibandingkan dengan sertifikat yang mencakup lebih banyak aspek legal seperti IMB (Izin Mendirikan Bangunan) dan PBB (Pajak Bumi dan Bangunan).
- Sertifikat lain seperti **PPJB** (Perjanjian Pengikatan Jual Beli), **AJB** (Akta Jual Beli), dan **HGB** juga menunjukkan harga yang relatif lebih rendah.

Dari hasil ini, dapat disimpulkan bahwa sertifikat yang lebih lengkap atau memiliki izin tambahan seperti IMB dan PBB cenderung memberikan nilai properti yang lebih tinggi, sementara sertifikat yang lebih sederhana atau bersifat sementara cenderung dikaitkan dengan harga yang lebih rendah.
